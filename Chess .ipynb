```python
# using filtering to obtain the name of the best opening for white players
filt = (df['victory_status'] == 'mate') & (df['winner'] == 'white')
```

```python
df_white_winner = df[filt]
```

```python
# df_white_winner is a new dataframe containing only chess games in which the winner by checkmate was the white player
df_white_winner
```

```python
#using the values_counts method to obtain the name of the most used opening in chess games where white players won by checkmate
df_white_winner['opening_name'].value_counts()
```

#### the Scandinavian Defense: Mieses-Kotroc Variation is the most used opening in games where white players achieved victory by checkmate

#### procedure to find out the performance of white players vs black players

```python
# counting the total number of games played by white players and black players
total_white_games = df['white_id'].count()
total_black_games = df['black_id'].count()
# counting the total number of wins by white players and black players 
white_wins = (df['winner'] == 'white').sum()
black_wins = (df['winner'] == 'black').sum()
# calculating the win rate for both types of players
white_win_rate = white_wins / total_white_games * 100
black_win_rate = black_wins / total_black_games * 100
# calculating the average rating of both white and black players
avg_white_rating = df['white_rating'].mean()
avg_black_rating = df['black_rating'].mean()
```

```python
#creating a new dataframe using this filtered data
performance_df = pd.DataFrame({
    "Color": ["White", "Black"],
    "Games_played": [total_white_games, total_black_games],
    "Victories": [white_wins, black_wins],
    "Win_rate": [white_win_rate, black_win_rate],
    "Avg_rating": [avg_white_rating, avg_black_rating]
})
```

```python
# new dataframe comparing the performance of white players vs black players
performance_df
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# Crear figura con subgráficos
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de porcentaje de victorias
axes[0].bar(performance_df["Color"], performance_df["Win_rate"], color=['blue', 'black'])
axes[0].set_title("Win_rates")
axes[0].set_ylabel("Porcentaje (%)")

# Gráfico de rating promedio
axes[1].bar(performance_df["Color"], performance_df["Avg_rating"], color=['blue', 'black'])
axes[1].set_title("Avg_rating")
axes[1].set_ylabel("Avg_elo")

# Mostrar gráficos
plt.show()
```
